## Creating a Tensor


```python
# using empty
a = torch.empty(2,3)
print(a)
```

```python
# check type
type(a)
```

```python
# using zeros
torch.zeros(2,3)
```

```python
# using ones
torch.ones(2,3)
```

```python
# using rand
torch.rand(2,3)
```

```python
# without use of seed
torch.rand(2,3)
```

```python
# use of manual_seed
torch.manual_seed(100)
torch.rand(2,3)
```

```python
# using tensor
torch.tensor([[1,2,3],[4,5,6]])
```

```python
# other ways

# arange
print("using a range ->", torch.arange(0,10,2))

# using linspace
print("using linspace ->", torch.linspace(0,10,10))

# using eye
print("using eye ->", torch.eye(5))

# using full
print("using full ->", torch.full((3, 3), 5))
```
