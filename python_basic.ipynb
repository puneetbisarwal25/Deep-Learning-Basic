```python
import numpy as np

W = np.array([[1, 2]])
x = np.array([[3],
              [4]])

b = 1

y = np.dot(W, x) + b

print(y)
```

Simple NumPy Neural Network Example

```python
import numpy as np

# Input
x = np.array([1, 2])

# Weights
W = np.array([0.5, 0.8])

# Bias
b = 0.1

# Forward pass
y = np.dot(x, W) + b

print(y)
```

```python
print("X Shape = ", x.shape)
```

```python
print("W Shape = ", W.shape)
```
